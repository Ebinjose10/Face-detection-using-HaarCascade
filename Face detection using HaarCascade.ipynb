```python
import cv2
import matplotlib.pyplot as plt
```

```python
#reading image
image=cv2.imread("barca.jpg")
image=hsv=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#grayscalling
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
#detecting face and number of faces
haar_cascade = cv2.CascadeClassifier('haarCascade_facedetection.xml')
face_detect = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=7)

print(f'Number of faces found = {len(face_detect)}')
face_detect
```

```python
#Drawing rectangel over detected face
for (x,y,w,h) in face_detect:
    detected_face=cv2.rectangle(image, (x,y), (x+w,y+h), (0,255,0), thickness=2)

plt.imshow(detected_face)
```
